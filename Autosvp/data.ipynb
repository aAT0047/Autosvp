```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\1\Desktop\code_ppg\Autosvp\merged_data.csv")

df['Sample ID'] = df['Sample ID'].str.replace('Sample_', '').astype(int)

dff1 = pd.read_csv(r"C:\Users\1\Desktop\code_ppg\Autosvp\merged_dataf1.csv")
dff1 = dff1.rename(columns={'sample_name': 'Sample ID'})
dff1value = pd.read_csv(r"C:\Users\1\Desktop\code_ppg\Autosvp\F1.csv")
dff1value = dff1value.rename(columns={'sample_name': 'Sample ID'})
```

```python

# 根据Sample ID合并df和dff1
merged_df = pd.merge(df, dff1, on='Sample ID', how='inner')
# # 重新排列列
cols = merged_df.columns.tolist()  # 获取列名列表
new_cols_order = cols[:1] + cols[-2:] + cols[1:-2]  # 新的列顺序
merged_df = merged_df[new_cols_order]
merged_df=  pd.merge(merged_df, dff1value, on='Sample ID', how='inner')
merged_df = merged_df[merged_df['f1_score'] > 0.74]
# 查看合并后的DataFrame
print(merged_df)

merged_df.to_csv(r'C:\Users\1\Desktop\code_ppg\Autosvp\finaF1.csv', index=False)
```
